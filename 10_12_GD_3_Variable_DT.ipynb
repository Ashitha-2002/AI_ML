```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 500) 
Y = X ** 2 - 25
```

```python
plt.plot(X,Y)
```

```python
# Building the model
beta0 = 0
beta1 = 0
beta2 = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = beta0 + beta1 * X + beta2 * X ** 2 # The current predicted value of Y
    D_beta0 = (-2/n) * sum(Y - Y_pred)  # Derivative wrt beta0
    D_beta1 = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt beta1
    D_beta2 = (-2/n) * sum(X**2 * (Y - Y_pred))  # Derivative wrt beta2
    beta0 = beta0 - L * D_beta0  # Update beta0
    beta1 = beta1 - L * D_beta1  # Update beta1
    beta2 = beta2 - L * D_beta2  # Update beta2
    
print ("beta0 = ",beta0)
print ("beta1 = ",beta1)
print ("beta2 = ",beta2)
```

```python
# Making predictions
Y_pred = beta0 + beta1 * X + beta2 * X ** 2

plt.plot(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```

```python
p = np.linspace(0, 1, 500)

# Compute entropy, handling log(0) by using np.where
entropy_p = -(np.where(p > 0, p * np.log(p), 0) + np.where(1-p > 0, (1-p) * np.log(1-p), 0))

```

```python
plt.figure(figsize=(8, 6))
plt.plot(p, entropy_p, label='Entropy H(p)', color='blue')
plt.title('Graph of Entropy H(p)')
plt.xlabel('p')
plt.ylabel('H(p)')
```
